```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from skimage.exposure import equalize_hist
```

```python
myImage = imread('abgefilmt.jpg')
```

```python
# Forward Fourier

dark_image_grey = rgb2gray(myImage)

# Orig
plt.imshow(dark_image_grey, cmap='gray')
plt.show()

# Fourier
dark_image_grey_fourier = np.fft.fftshift(np.fft.fft2(dark_image_grey))
plt.imshow(np.log(abs(dark_image_grey_fourier)), cmap='gray')
plt.show()


print(dark_image_grey_fourier[0][0])
print(abs(dark_image_grey_fourier[0][0]))
# plt.imsave("dark_image_grey_fourier.png", np.log(abs(dark_image_grey_fourier)))
# dark_image_grey_fourier = plt.imread("dark_image_grey_fourier.png")

# Inverse-Fourier
plt.imshow(abs(np.fft.ifft2(dark_image_grey_fourier)), cmap='gray')
plt.show()

```

```python
# Inverse Fourier


fourier_image = imread("dark_image_grey_fourier.jpg")

fourier_image = rgb2gray(fourier_image)

plt.imshow(np.log(abs(np.fft.ifft2(fourier_image))),cmap='gray')
```
